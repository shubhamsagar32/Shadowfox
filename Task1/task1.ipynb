```python
dataset.fillna(dataset["CRIM"].mean(),inplace=True)
dataset.fillna(dataset["ZN"].mean(),inplace=True)
dataset.fillna(dataset["INDUS"].mean(),inplace=True)
dataset.fillna(dataset["CHAS"].mean(),inplace=True)
dataset.fillna(dataset["AGE"].mean(),inplace=True)
dataset.fillna(dataset["LSTAT"].mean(),inplace=True)

dataset
```

```python
plt.figure(figsize=(22,8))
plt.scatter(dataset.index,dataset["MEDV"],color="red",lw=3)
plt.plot(dataset.index,dataset["MEDV"])
plt.title("Median value of homes (in $1000)")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
fig, ax1 = plt.subplots(7,2, figsize=(20,25))
k = 0
columns = list(dataset.columns)
for i in range(7):
    for j in range(2):
            sns.histplot(dataset[columns[k]], ax = ax1[i][j], color = 'green')
            ax1[i][j].grid(True)
            k += 1
plt.show()
```

```python
def log_transform(col):
    return np.log(col[0])

dataset["DIS"]=dataset[["DIS"]].apply(log_transform, axis=1)

sns.histplot(dataset["DIS"], color = 'green')
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

X=dataset.iloc[:,0:13]
Y=dataset.iloc[:,13]

print("Unscaled Data: \n")
display(X)  
scaler = StandardScaler()
X = scaler.fit_transform(X)
print("Scaled Data: \n")
display(X)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=94)
```

```python
import xgboost as xgb
xgbr = xgb.XGBRegressor(objective='reg:squarederror',
                        random_state=50,
                        max_depth=6,
                        learning_rate = 0.08,
                        n_estimators = 500,
                        colsample_bylevel = 0.4,
                        reg_alpha = 1,
                        subsample = 1,
                       )

xgbr.fit(X_train,Y_train)
Y_pred = xgbr.predict(X_test)
mse=mean_squared_error(Y_test, Y_pred)
score=r2_score(Y_test,Y_pred)
print("Root Mean Square Error: %.2f" % (mse**(0.5)))
print("Accuracy: {} %".format(round((score*100),2)))
```
